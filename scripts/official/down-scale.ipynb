```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("originals/00.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(5,5))
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

```python
img_bis = img[::4,::4,:]
plt.figure(figsize=(5,5))
plt.imshow(img_bis)
plt.axis("off")
plt.title("DownScaled Image")
plt.show()
```
